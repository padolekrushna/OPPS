subplots creates a figure and a grid of subplots with a single call, while providing reasonable control over how the individual plots are created.

```python
 #temperature and week day
import pandas as pd
data = {'Days': ['Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday'],
        'Week 1': [27,29,32,33,33,31,29],
        'Week 2': [29,30,31,33,34,34,31],
        'Week 3': [29,28,25,27,31,30,29]}

days = data['Days']
week1 = data['Week 1']
week2 = data['Week 2']
week3 = data['Week 3']

df = pd.DataFrame(data)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline   #display matplotlib plots inline within the notebook
#Plotting for Week 1
fig, ax = plt.subplots()   #creates a new figure and fig represents the entire figure or canvas, and ax represents the axes of a single subplot.
ax.plot(days, week1)
ax.set_title('A single plot')
```

```python
"""
Adding elements to the subplot

Use ax.set_title() for setting a plot title
Use ax.set_xlabel() and ax.set_ylabel() for labeling x and y-axis respectively
Use ax.legend() for the observation variables
"""
fig, ax = plt.subplots()
ax.plot(days, week1)
ax.set_xlabel('Days')
ax.set_ylabel('Temperature')
ax.set_title('A single plot')
ax.legend(['Week 1'])
```

```python
#stacking subplots vertically
fig, axs = plt.subplots(3, figsize=(10,5))  #This line creates a figure with 3 subplots stacked vertically. The figsize=(10,5) parameter sets the size of the figure to be 10 inches wide and 5 inches tall.
fig.suptitle('Vertically stacked subplots')
axs[0].plot(days, week1, color = 'red')
axs[1].plot(days, week2, color = 'orange')
axs[2].plot(days, week3, color = 'green')
```

```python
#horizontal stacking
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,3))
fig.suptitle('Horizontally stacked subplots')
ax1.plot(days, week1)
ax2.plot(days, week2, color = 'orange')
ax3.plot(days, week3, color = 'green')
```

```python
#stacking in two directions
fig, ax = plt.subplots(2, 2, figsize=(15,4))
fig.suptitle('Subplots in two directions')
ax[0][0].plot(days, week1)
ax[0][1].plot(days, week2, color = 'orange')
ax[1][0].plot(days, week3, color = 'green')
```
